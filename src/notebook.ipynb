```python
def post_questao1(image_path):
    url = 'http://localhost:5000/questao1'
    with open(image_path, 'rb') as image_file:
        files = {'image': image_file}
        response = requests.post(url, files=files)

    if response.status_code == 200:
        result = response.json()
        return result

    else:
        print('Erro ao enviar a imagem')
        
result = post_questao1('/home/v/Desktop/teste-de-visao/src/img46_gray_noise.png')
img = decode_base64_image(result['result'])
cvt_and_plot(img)
```
