```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
def read_img(path):
   return cv2.imread(path)
noisy_image = read_img('img_q_1/img46_gray_noise.png')
plt.imshow(noisy_image)
plt.axis('off')
plt.show()
```

```python
filtered_image = cv2.medianBlur(noisy_image, 3)
plt.imshow(filtered_image)
plt.axis('off')
plt.show()
```

```python
fig, axes = plt.subplots(3, 3, figsize=(8, 8))
images = []
lista = [3, 5, 7, 9, 11, 13, 15, 17, 19]

for i in lista:
    images.append(cv2.medianBlur(noisy_image, i))
for i, ax in enumerate(axes.flat):
    ax.imshow(images[i])
    ax.axis('off')

plt.tight_layout()
plt.show()
```
