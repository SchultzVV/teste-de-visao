```python
import numpy as np
import cv2

# img = cv2.imread('img_q_2/img67.png')
gray = cv2.imread('img_q_2/img67.png', 0)
template = cv2.imread('img_q_2/silo.png', 0)
h, w = template.shape

methods = [cv2.TM_CCOEFF, cv2.TM_CCOEFF_NORMED, cv2.TM_CCORR,
            cv2.TM_CCORR_NORMED, cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]

for method in methods:
    img2 = gray.copy()

    result = cv2.matchTemplate(img2, template, method)
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
    if method in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
        location = min_loc
    else:
        location = max_loc
    img = cv2.imread('img_q_2/img67.png')
    bottom_right = (location[0] + w, location[1] + h)    
    cv2.rectangle(img, location, bottom_right, 255, 5)
    cv2.imshow('Match', img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
def find_silo(path):
    img = cv2.imread(path)
    hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

    lower = np.array([0, 0, 41])
    uper= np.array([20, 89, 73])
    mask2 = cv2.inRange(hsv_img, lower, uper)
    lower = np.array([0, 10, 43])
    uper= np.array([16, 124, 93])
    # lower = np.array([0, 0, 51])
    # uper= np.array([13, 147, 81])
    lower = np.array([0, 10, 25])
    uper= np.array([16, 140, 84])
    mask2 = cv2.inRange(hsv_img, lower, uper)
    det_obj = cv2.bitwise_and(img, img, mask=mask2)

    cn, h = cv2.findContours(mask2.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if len(cn) > 0:
        #for i in range(min(4, len(cn))):
        sorted_contours = sorted(cn, key=cv2.contourArea, reverse=True)[:5]
        for contour in sorted_contours:
            area = cv2.contourArea(contour)
            if area > 60:
                x, y, w, h = cv2.boundingRect(contour)
                cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 3)
                print('Object detected')
    return img

img1 = find_silo('img_q_2/img46.png')
img2 = find_silo('img_q_2/img47.png')
img3 = find_silo('img_q_2/img66.png')
img4 = find_silo('img_q_2/img67.png')

cv2.imshow('Match1', img1)
cv2.imshow('Match2', img2)
cv2.imshow('Match3', img3)
cv2.imshow('Match4', img4)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import mahotas
import numpy as np
import cv2
img = cv2.imread('img_q_2/img46.png', 0)
# img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) # converte
suave = cv2.GaussianBlur(img, (7, 7), 0) # aplica blur
T = mahotas.thresholding.otsu(suave)
temp = img.copy()
temp[temp > T] = 255
temp[temp < 255] = 0
temp = cv2.bitwise_not(temp)
T = mahotas.thresholding.rc(suave)
temp2 = img.copy()
temp2[temp2 > T] = 255
temp2[temp2 < 255] = 0
temp2 = cv2.bitwise_not(temp2)
resultado = np.vstack([np.hstack([img, suave]),np.hstack([temp, temp2])])
cv2.imshow("Binarizao com mtodo Otsu e Riddler-Calvard",resultado)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
